```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Split Train and Test in Time Dimension

```python
path = './result123/loss.txt'
loss = np.loadtxt(path)
end_epoch = loss.shape[0]
train_loss = loss[:, 1]
test_loss = loss[:, 2]


print(end_epoch)
plt.plot(train_loss,'--', label = 'train_loss')
plt.plot(test_loss, label = 'test_loss')
plt.legend()
plt.ylim(0.0, 0.15)
plt.xlabel('epochs')
plt.ylabel('MSE Loss')
plt.show()
plt.close()
```
