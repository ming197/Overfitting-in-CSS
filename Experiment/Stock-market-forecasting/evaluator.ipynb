```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

scores = pd.read_csv("loss(1993).csv", index_col = "epochs")
loss_columns = ['train loss','val loss']
acc_columns = ['train acc','val acc']

end_epoch = scores.shape[0]

plt.plot(scores[loss_columns[0]][:end_epoch],'--', label = 'Train Loss')
plt.plot(scores[loss_columns[1]][:end_epoch], label = 'Val Loss')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('Cross Entropy')
plt.show()
plt.close()
plt.plot(scores[acc_columns[0]][:end_epoch],'--', label = 'Train Acc')
plt.plot(scores[acc_columns[1]][:end_epoch], label = 'Val Acc')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('Accuracy')
plt.show()
plt.close()
```
